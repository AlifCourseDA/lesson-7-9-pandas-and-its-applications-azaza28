````python
fig = sns.barplot(x = df.transaction.unique(), 
            y = df.transaction.value_counts())
fig.set(xlabel='Transaction Status', ylabel='Occurences')
plt.show()
````

I guess, that we could have groupped the data by name, with the help of `hue` parameter and visualize the distribution of transactions per user.

````python
forhist = df.query("transaction == 'successfull'").groupby(['name', 'transaction']) \
.agg({'transaction':'count'}).rename({'transaction':'occurence'}, axis = 1) \
.sort_values('occurence', ascending = False).reset_index().drop('transaction',axis = 1)
````

````python
plt.hist(x = forhist['occurence'], edgecolor='black')
````

### In the above cell we have devided our users into 10 subgroups (bins). However, if you want to take a glance at the more descriptive graph - Enjoy the cell that in below üòâ

````python
plt.figure(figsize=(15, 10))
values, bins, bars = plt.hist(x = forhist['occurence'], bins = 56, color = 'pink')
plt.bar_label(bars, fontsize = 15, color='green')
plt.xlabel("Number of Transactions", fontsize = 18)
plt.ylabel("Number of People w/ same number of transactions", fontsize = 18)
plt.title('Successful Transactions\' Distrubtion', fontsize = 18)
plt.margins(x=0.01, y=0.1)
plt.show()
````

````python
germany_top = retail.query('Country == "Germany"') \
.groupby("CustomerID") \
.agg({'InvoiceNo':'count'}).reset_index()
germany_top =germany_top[germany_top['InvoiceNo'] > germany_top['InvoiceNo'].quantile(0.8)]['CustomerID']
germany_top
````

As we already got the most active germans, let's get all the information about them from our DataFrame

````python
top_retail_germany = retail[retail.CustomerID.isin(germany_top)]

top_retail_germany[top_retail_germany['StockCode'] != 'POST'] \
.groupby('StockCode') \
.agg({'Quantity':'count'}) \
.sort_values('Quantity',ascending = False).reset_index()
````

```python3
top_retail_germany[top_retail_germany['StockCode'] != 'POST'] \

.merge(retail[['StockCode','Description']])
```

````python
retail['Revenue'] = retail.UnitPrice * retail.Quantity
````

> In the above cell we have calculated and created the `Revenue` column, as we have been asked.

````python
top = []
print('The 5 most expensive deals (in descending order):')
for index, i in enumerate (retail.groupby('InvoiceNo') \
.agg({'Revenue':'sum'}).sort_values('Revenue',ascending=False) \
.reset_index().rename({'Revenue':'TotalRevenue'},axis = 1).head()['InvoiceNo']):
    if index == 4:
        print(i)
    else:
        print(i, end=', ')
    top.append(i)
````
